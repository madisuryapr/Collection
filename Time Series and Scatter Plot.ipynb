```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style = "darkgrid")
```

```python
Payments = pd.read_excel("Modeling_Dataset.xlsx")
Payments
```

# Linear Regression Scatter Plot for Industrial Productivity vs Noncash Payments

```python
# IPI vs Transactions Value (N) of Debit and Credit Cards
sns.pairplot(data = Payments, x_vars = ["NDEBIT", "NCREDIT"],
             y_vars = ["IPI"], height = 5, aspect = .8, kind = "reg");
```

```python
# IPI vs Transactions Value (N) of Clearing and RTGS System
sns.pairplot(data = Payments, x_vars = ["NCLEARING", "NRTGS"],
             y_vars = ["IPI"], height = 5, aspect = .8, kind = "reg");
```

```python
# IPI vs Transactions Volume (V) of Debit and Credit Cards
sns.pairplot(data = Payments, x_vars = ["VDEBIT", "VCREDIT"],
             y_vars = ["IPI"], height = 5, aspect = .8, kind = "reg");
```

```python
# IPI vs Transactions Volume (V) of Clearing and RTGS System
sns.pairplot(data = Payments, x_vars = ["VCLEARING", "VRTGS"],
             y_vars = ["IPI"], height = 5, aspect = .8, kind = "reg");
```

Time Series Plot for corresponding Variables

```python
# Time series plot for Transactions Value of Debit Cards
Debits_Value = sns.relplot(data = Payments, x = "Date", y = "NDEBIT",
                          kind = "line", aspect = 1.25);
Debits_Value.figure.autofmt_xdate()
plt.savefig("Debits_Value.png", dpi = 450)
```

```python
# Time series plot for Transactions Value of Credit Cards
Debits_Value = sns.relplot(data = Payments, x = "Date", y = "NCREDIT",
                          kind = "line", aspect = 1.25);
Debits_Value.figure.autofmt_xdate()
plt.savefig("Credits_Value.png", dpi = 450)
```
